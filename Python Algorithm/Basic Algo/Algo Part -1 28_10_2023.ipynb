```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'id':[1,2,3,4,5,6,7],'name':['A','B','C','D','E','F','G']})
print(df1)
```

```python
df2 = pd.DataFrame({'id':[3,6,7],'marks':[200,256,180]})
print(df2)
```

```python
print("Inner Join")
df3 = df1.merge(df2,on = 'id',how = 'inner')
print(df3)
```

```python
print("Left Join")
df4 = df1.merge(df2,on = 'id',how = 'left')
print(df4)
```

```python
print("Right Join")
df5 = df1.merge(df2,on = 'id',how = 'right')
print(df5)
```

```python
print("Full Outer Join")
df6 = df1.merge(df2,on = 'id',how = 'outer')
print(df6)
```

```python
##### Simple Linear Regression  ------- 
import seaborn as sns
```

```python
#sns.get_dataset_names()
```

```python
data = sns.load_dataset('mpg')
print(data.info())
```

```python
data1 = data[['mpg', 'weight']]
data1
```

```python
sns.regplot(x = 'weight', y = 'mpg', data = data1)
```

```python
X = data1['weight'].values
X = X.reshape(-1,1)   ####### It applies only on simple linear regression
print(X)
```

```python
y = data1['mpg'].values
y = y.reshape(-1,1)   ####### It applies only on simple linear regression
print(y)
```

```python
##### Importing Scikit Learn for Algorithgms
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
print(model)
```

```python
%%time
#### Training the Model/Sytem  - By Fitting my data
model_fit = model.fit(X,y)
print(model_fit)
```

```python
model_fit.coef_
```

```python
model_fit.intercept_
```

```python
y = -0.00767661*weight + 46.31736442
print(y)
```

```python
predict = model_fit.predict(X)  #### Will be done on test data for validation and after that raw data
predict
```
